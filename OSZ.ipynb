```python
import numpy as np
import matplotlib.pyplot as plt
```

# Messwerte

```python
u = np.array([8.01,5.47,3.71,2.55,1.75,1.22,0.82,0.58,0.41,0.27,0.17])
t = np.arange(0,2001,200.0)

#u = np.array([8.00,7.00,6.00,5.46,5.00,3.7,3.00,2.48,1.66,1.09,0.713])
#t = np.array([0.0,0.069,0.15,0.2,0.245,0.4,0.507,0.6,0.8,1.0,1.2])

umax =  u+0.05
umin =  u-0.05

print(u)
print(t)
```

# Fit

```python
coefficents = np.polyfit(t,np.log(u),1)
fit = np.poly1d(coefficents)
fit_graph = fit(t)

maxcoefficents = np.polyfit(t,np.log(umax),1)
maxfit = np.poly1d(maxcoefficents)
maxfit_graph = maxfit(t)

mincoefficents = np.polyfit(t,np.log(umin),1)
minfit = np.poly1d(mincoefficents)
minfit_graph = minfit(t)

delta_Steigung = maxcoefficents[0]-mincoefficents[0]

print('Die maximale steigung ist '+str(np.round(maxcoefficents[0],5)))
print('Die minimale steigung ist '+str(np.round(mincoefficents[0],5)))

print('Die Steigung ist: '+str(np.round(coefficents[0],5))+'±'+str(np.round(delta_Steigung,4))+' ln(V)/ms')
```

# Plot

```python
plt.figure(figsize = (11,6))
plt.title('Teilversuch 5')
plt.xlabel('Zeit [ms]')
plt.ylabel('Ladung [V]')
plt.plot(t,np.log(u),'o')
plt.plot(t,fit_graph)
plt.plot(t,minfit_graph)
plt.plot(t,maxfit_graph)
```

# Relaktionszeit

```python
tau = -1/coefficents[0]
taumax = -1/maxcoefficents[0]
taumin = -1/mincoefficents[0]
delta_tau = taumax-taumin


print('Relaktion findet in '+str(np.round(tau,-1))+'±'+str(np.round(delta_tau,0))+' ms statt.')
```
