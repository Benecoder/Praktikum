```python
import numpy as np
import matplotlib.pyplot as plt
```

# Teilversuch 2
### gemmessene Werte

```python
alpha = np.linspace(0,90,10)
beta = np.array([148.0,172.5,194.5,216.0,234.0,
                 251.5,269.5,285.5,298.0,318.5])
beta = beta-beta[0]
```

Torsionswinkel phi

```python
phi = beta-alpha
```

### Line Fit

```python
coefficents = np.polyfit(phi,np.sin(np.radians(alpha)),1)
print('Abstand zum Ursprung ist: '+str(np.round(np.abs(coefficents[1]),6))+'.')

fit = np.poly1d(coefficents)
fit_graph = fit(phi)
```

### Plot

```python
plt.figure(figsize=(10,5))
plt.title('Messung Teilversuch 2')
plt.xlabel('Torsionswinkel [°]')
plt.ylabel('sin(α)')


plt.errorbar(phi,np.sin(np.radians(alpha)),yerr=0,xerr=2,fmt='.')
plt.plot(phi,fit_graph,'red')

plt.show()
```

# Teilversuch 4
### gemessene Werte

```python
Uind = np.array([0.164,0.145,0.108,0.071])
I = np.array([0.726,0.651,0.484,0.292])
```

### Line Fit

```python
coeff = np.polyfit(Uind,I,1)
fit = np.poly1d(coeff)
fit_graph = fit(Uind)
```

### Plot

```python
plt.figure(figsize=(10,5))
plt.title('Teilversuch 4')
plt.errorbar(Uind,I,xerr = 0.05*0.025,yerr=0.05*0.167,fmt='.')
plt.plot(Uind,fit_graph,'red')
plt.xlabel('Induzierte Spannung [V]')
plt.ylabel('Induzierter Strom [A]')
plt.show()
```
