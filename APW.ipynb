# Teilversuch 5
### Messreihe

```python
t = np.array([80,100,130,160,190,210,240,270,300,330,350])+273.15
u = np.array([0.016,0.022,0.034,0.055,0.078,0.095,0.127,0.159,0.202,0.252,0.291])
t0 = 29.7+273.15

t_error = 0.1
t0_error = 0.5
u_error = 0.002

u_min = u-u_error
u_max = u+u_error

t4_t04 = t**4-t0**4
t4_t04_min = (t-t_error)**4-(t0+t0_error)**4
t4_t04_max = (t+t_error)**4-(t0-t0_error)**4
```

### Curve_fit

```python
coeff = np.polyfit(t4_t04,u,1)
coeff_min = np.polyfit(t4_t04_min,u_min,1)
coeff_max = np.polyfit(t4_t04_max,u_max,1)

fit = np.poly1d(coeff)
fit_min = np.poly1d(coeff_min)
fit_max = np.poly1d(coeff_max)

t_coords = np.arange(0,1.5e11,1e9)
fit_graph  = fit(t_coords)
fit_graph_min = fit_min(t_coords)
fit_graph_max = fit_max(t_coords)
```

```python
plt.figure(figsize=(20,9))
plt.title('StrahlungsintensitÃ¤t eines Schwarzraumstrahlers')
plt.ylabel('Thermospannung [mv]')
plt.xlabel('T^4-T0^4 [K^4]')
plt.plot(t4_t04,u,'o')
plt.plot(t_coords,fit_graph,color = 'black')
plt.plot(t_coords,fit_graph_min,color = 'black',alpha = 0.5)
plt.plot(t_coords,fit_graph_max,color = 'black',alpha = 0.5)
plt.grid(True)
plt.show()
```
